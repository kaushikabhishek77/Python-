# Task 2
## Filling missing values in Cost column of project_df with running average of preceding values.

```python
# Wrting a function to impute missing values with running average

def imputing_func(series):

    series_copy = series.copy()
    
    for i,j in enumerate(series_copy):
        if pd.isna(series_copy[i]):
            series_copy[i] = np.mean(series_copy[:i])
    
    return series_copy
```

```python
# Applying the function written just above to impute missing values in Cost column with running average

project_df['Cost'] = imputing_func(project_df['Cost'])
project_df
```

# Task 3 
## Splitting Name into two disparate columns - First Name and Last Name and dropping Name column in employees_df

```python
employees_df[['First Name','Last Name']] = employees_df['Name'].str.split(' ', expand = True)  # 'expand' parameter should be set to True for creating separate column
employees_df = employees_df.drop(['Name'], axis=1)                         # axis = 1 for dropping column
```

```python
employees_df
```

# Task 4 
## Joining all dataframes to create Final dataframe

```python
merged_df_1 = pd.merge(left= employees_df, right= seniority_df, on = 'ID') # Using merge function of Pandas library
Final = merged_df_1.merge(project_df, on = 'ID')                           # Using merge method of dataframe object 
```

```python
Final
```

# Task 5 
## Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects

```python
Final['Bonus'] = Final[Final['Status'].isin(['Finished'])]['Cost'] * 0.05  ## Creating a new 'bonus' column for 'Finished' projects
Final['Bonus'] = Final['Bonus'].replace(np.NaN, 0)                         ## Replacing Null values with 0
Final
```

# Task 6
## Demote the designation level by 1, whose projects have status fail. Delete the employees record whose designation level is above 4

```python
# Creating a function to demote repsective Designation Levels for those with 'Failed' projects

def temp(df):

    if df['Status'] == 'Failed':
        df['Designation Level'] = df['Designation Level'] + 1
    
    return df
```

```python
Final = Final.apply(temp, axis=1)     # Using 'apply' method to dataframe and axis = 1 because the operation has to
                                      # be conducted on another column in same row.
```

```python
Final = Final[Final['Designation Level'] != 4]  # Excluding employees with Designation Level of 4.
Final.reset_index(drop=True, inplace=True)      # Resetting index and dropping the newly created 'index' column and setting inplace = True  
Final
```

# Task 7
## Add Mr. and Mrs. to the first name column and drop the gender column.

```python
# Creating a function to add salutation to the first name column based .
# The function woould exploit the df.iterrows() method to transform the First Name column

def salutation_func(df):
    
    for lab, row in df.iterrows():
        if row['Gender'] == 'M':
            df.loc[lab, 'First Name'] = 'Mr. ' + df.loc[lab, 'First Name']
        elif row['Gender'] == 'F':
            df.loc[lab, 'First Name'] = 'Mrs. ' + df.loc[lab, 'First Name']
        else:
            continue
    
    return df 
```

```python
Final = salutation_func(Final)
```

```python
# Dropping 'Gender' column and using inplace = True so that the changes are made in existing part of memory 
# without the creation of new dataframe.

Final.drop(['Gender'],axis=1, inplace=True) 
```

```python
Final
```

# Task 8
## Promote designation level by 1 for the employees whose age is more than 29 years using IF condition

```python
# Using 'for' loop and df.iterrows() method to iterate through each row and promoting an employee more than 29 years of age.

for lab, row in Final.iterrows():       
    if row['Age'] > 29:
        Final.loc[lab,'Designation Level'] = Final.loc[lab,'Designation Level'] - 1
```
